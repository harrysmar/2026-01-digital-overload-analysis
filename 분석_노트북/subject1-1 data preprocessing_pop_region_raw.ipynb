```python
## 행정구역별 인구수 합계 데이터 전처리
df = pop_region_raw.copy()

# 2024열의 모든 값을 숫자로 변환하고 숫자로 바꿀 수 없는것은 NaN(결측치)로 처리
df["2024"] = pd.to_numeric(df["2024"], errors = "coerce")

# 2024열이 결측치인 행 제거
df = df.dropna(subset=["2024"])

# 2024열의 값들을 소수점 없이 정수로 맞춤
df["2024"] = df["2024"].astype(int)

df.head()
```

```python
# 행정구역 1레벨 이름 목록 리스트
sido = ["서울특별시", "부산광역시", "대구광역시", "인천광역시", "광주광역시", "대전광역시", "울산광역시", "세종특별자치시",
        "경기도", "강원특별자치도", "충청북도", "충청남도", "전북특별자치도", "전라남도", "경상북도", "경상남도", "제주특별자치도"]

# sido 리스트에 들어있는 값이면 그대로 두고, 아니면 NaN으로 바꿈
# 원본값.where(조건) 조건이 TRUE인 칸은 원래 값 유지, 조건이 FALSE인 칸은 NaN으로 바꿈
df["시도"] = df["행정구역별(읍면동)"].where(df["행정구역별(읍면동)"].isin(sido))

# 시도 열의 자료형을 문자열로 확정한 다음, 위에서 아래로 내려가며 NaN을 바로 위의 값으로 채움
df["시도"] = df["시도"].astype("string").ffill()
df.head()
```

```python
# 행정구역별(읍면동) 열의 값이 sido 리스트에 포함된 행은 TRUE가 되는데 앞에 ~를 붙여 FALSE로 만들고 TRUE인 행만 남김
df = df[~df["행정구역별(읍면동)"].isin(sido)].copy()
df.head(100)
```

```python
# 세종특별자치시는 행정구역 2레벨이 없는 예외라서 데이터 전처리가 잘 되었는지 다시 한번 확인
df[df["행정구역별(읍면동)"].str.contains("세종", na=False)]
```

```python
# 행정구역 열을 새로 만들어서 정확한 행정구역 명칭을 저장
df["행정구역"] = df["시도"] + " " + df["행정구역별(읍면동)"]
df.head(308)
```

```python
# 행정구역별(읍면동)열의 값이 읍부, 면부, 동부인 행은 TRUE가 되는데 
# 앞에 ~를 붙여 FALSE(읍부, 면부, 동부가 이에 해당)로 만들고 TRUE인 값이 포함된 행만 남김
df = df[~df["행정구역별(읍면동)"].isin(["읍부", "면부", "동부"])].copy()

# 필터링 후 인덱스가 중간중간 비어있는 것을 방지하기 위해 0부터 다시 연속 번호로 인덱스를 재설정 
df = df.reset_index(drop=True)

df.head(266)
```

```python
# 필요없는 열 삭제, 이름이 헷갈리는 열 이름 바꾸기
df = df.drop(columns = ["행정구역별(읍면동)", "연령별", "시도"])
df = df.rename(columns={"2024":"인구 수"})
```

```python
df.head(300)
```

```python
# 행정구역 열에서 NaN이 있어도 문자열로 처리하고 앞뒤 공백을 제거한 후에 문자열이 시, 군, 구로 끝나면 TRUE 아니면 FALSE
mask_end_sigungu = df["행정구역"].astype(str).str.strip().str.endswith(("시","군","구"))

# 앞에 ~을 붙여서 시, 군, 구로 끝나지 않으면 TRUE 아니면 FALSE로 한다음 TRUE인 값만 not_sigungu에 저장
not_sigungu = df[~mask_end_sigungu].copy()
not_sigungu =not_sigungu.reset_index(drop=True)
not_sigungu.head() 

# 행정구역명의 끝이 시, 군, 구로 끝나서 중복되는 행은 없음
# 그렇다면 왜 264개인가? -> 포항시 북구, 창원시 성산구가 따로따로 집계되어 226개보다 더 늘어난것으로 보임
```

```python
# 행정구역 열을 전부 문자열로 바꾸고 "\xa0"으로 안깨지는 특수 공백을 일반 공백으로 바꾸고 
# "r"\s+"로 공백 여러개를 하나로 압축하고 .str.strip()로 문자열 앞뒤 공백을 제거함
# 결과적으로 district는 행정구역 정리본임
district = (df["행정구역"].astype(str)
     .str.replace("\xa0", " ", regex=False)
     .str.replace(r"\s+", " ", regex=True)
     .str.strip())

# district 문자열들을 split()을 사용하여 공백 기준으로 나누어 리스트를 만듦
parts = district.str.split()

# parts.str[0]을 사용하여 각 리스트의 첫 단어를 가져오고 parts.str[-1]을 사용하여 각 리스트의 마지막 단어를 가져옴
first = parts.str[0]
last  = parts.str[-1]

# 리스트의 첫 단어가 "도" 또는 "특별자치도" 이고 리스트의 마지막 단어가 "구"인 리스트를 삭제대상으로 만듦
mask_do = first.str.endswith(("도","특별자치도")) & last.str.endswith("구")

print("삭제된 행 개수:", mask_do.sum())

# 앞에 ~을 붙여서 "도"|"특별자치도" + "구"인 문자열들을 FALSE로 만든 다음 삭제
df = df[~mask_do].copy()
df = df.reset_index(drop=True)
df.head(300)

# 시,군,구 개수가 229개인 이유는 226개에서 세종특별자치시와 제주특별자치도 제주시, 서귀포시를 포함한 값임
```

```python
# 전처리된 데이터 저장
df.to_csv("../4_데이터_전처리_자료/시,군,구별 전체 인구 합계 최종.csv", index=False, encoding="utf-8-sig")
```
