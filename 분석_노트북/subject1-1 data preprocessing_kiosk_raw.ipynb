```python
import pandas as pd

kiosk_raw = pd.read_csv("../3_데이터_자료/무인민원발급기 설치 정보.csv")
kiosk_raw.head()
```

```python
df = kiosk_raw.copy()

# 필요한 열만 선택
df = df[["관리기관명"]].copy()
df.head(5780)
```

```python
# 관리기관명 컬럼을 문자열로 통일하고, 깨진 공백/중복 공백 정리한 뒤 양끝 공백 제거
df["관리기관명"] = (df["관리기관명"].astype(str)
                 .str.replace("\xa0", " ", regex=False)
                 .str.replace(r"\s+", " ", regex=True)
                 .str.strip())

# 각 행을 1로 두고, 나중에 행정구역별로 합쳐서 설치 대수를 집계
df["설치대수"] = 1

# 행정구역별로 설치대수(1들의 합)를 집계해서 행정구역별 설치 건수 테이블로 변환
df = df.groupby("관리기관명", sort=False)["설치대수"].sum().reset_index()
df.head()
```

```python
df = df.rename(columns={"관리기관명":"행정구역"})
df.head(229)
```

```python
# 전처리된 데이터 저장
df.to_csv("../4_데이터_전처리_자료/무인민원발급기 설치 정보 최종.csv", index=False, encoding="utf-8-sig")
```
